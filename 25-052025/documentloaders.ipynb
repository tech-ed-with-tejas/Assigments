```python
## CHUNKING/Text splitter
```

```python
loader=PyPDFLoader('testpdf.pdf')
docs=loader.load()
text_splitter=RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=100)
final_documents=text_splitter.split_documents(docs)
```

```python
for i in final_documents[:4]:
    print(i.page_content)
    print("next chunk")
```

```python
text_splitter=CharacterTextSplitter(separator="\n\n",chunk_size=300,chunk_overlap=50)
splited_doc = text_splitter.split_documents(docs)
```

```python
for i in splited_doc[:4]:
    print(i.page_content)
    print("next chunk")
```
